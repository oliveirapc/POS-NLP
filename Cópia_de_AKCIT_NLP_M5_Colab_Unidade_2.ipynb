**Objetivo:** Compreender e aplicar o processo de lematizao.

***Tarefa 1.1:*** Utilizar a biblioteca spacy para lematizar um conjunto de palavras fornecidas.

**Definio de lemetizao:**

A lematizao  o processo de reduzir palavras s suas formas base ou raiz, conhecidas como lemas. Esse processo leva em conta o contexto em que a palavra  usada, transformando diferentes formas gramaticais em uma nica representao. Isso  particularmente til em tarefas de processamento de linguagem natural porque ajuda a unificar palavras que tm a mesma raiz, mas diferentes formas.

**Exemplo:** Vamos considerar as seguintes palavras em portugus: **correndo, correu, correm, muitos, justos**

Ao aplicar a lematizao, essas palavras so reduzidas s suas formas base: **correndo -> correr, correu -> correr, correm -> correr, muitos -> muito, justos -> justo**

**Nota:** Neste exemplo, a biblioteca SpaCy  utilizada para processar o texto e aplicar a lematizao. As palavras "correndo", "correu" e "correm" so reduzidas ao lema "correr", enquanto "muitos" e "justos" so simplificadas para "muito" e "justo", respectivamente.

A seguir, um pequeno exemplo de como realizar a lematizao em Python usando a biblioteca SpaCy:

1.   execute o cdigo e analise a sada; e

**Definio: Etiquetagem de partes do discurso (ou POS tagging):**

Trata-se de uma tcnica no processamento de linguagem natural que consiste em atribuir etiquetas gramaticais a cada palavra em um texto. Essas etiquetas identificam a funo gramatical das palavras, como substantivo, verbo, adjetivo, advrbio, entre outros.

**Abordagens Baseadas em Regras:**

Estas abordagens utilizam um conjunto de regras gramaticais manuais para atribuir etiquetas.

**Ferramentas e Bibliotecas:**
Diversas ferramentas e bibliotecas de PLN oferecem funcionalidades de POS tagging. Algumas das mais populares incluem:

**NLTK (Natural Language Toolkit):** Biblioteca poderosa em Python, oferece etiquetagem de POS junto com muitas outras funcionalidades para PLN.

**SpaCy:** Biblioteca em Python otimizada para desempenho e produo, com modelos pr-treinados para vrias lnguas.

**Legenda: Cores e Tags**

Substantivos (NN, NNS): White

Verbos (VB, VBD, VBZ, etc.): Verde

Adjetivos (JJ): Laranja

Advrbios (RB): Roxo

Determinantes e Preposies (DT, IN): Vermelho


**Segue exemplo de Cdigo com NLTK:**

1.   execute o cdigo e analise a sada; e
2.   mude as palavras da linha 14 do cdigo Python e re-execute.

```python
import subprocess
import sys

# Funo para instalar pacotes
def instalar_pacote(pacote, versao=None):
    subprocess.check_call([sys.executable, "-m", "pip", "install", pacote])

# Verificar e instalar NLTK se necessrio
try:
    import nltk
except ImportError:
    instalar_pacote("nltk")
    import nltk

from nltk.tokenize import word_tokenize

# Baixar o modelo de POS tagging
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')  # add
nltk.download('punkt')
nltk.download('punkt_tab')   # add

# Texto a ser etiquetado
texto = "The cat sleeps on the porch."
tokens = word_tokenize(texto)
pos_tags = nltk.pos_tag(tokens)

# Funo para colorir e formatar texto
def format_pos_tags(pos_tags):
    formatted_text = []
    for word, tag in pos_tags:
        color = "black"
        if tag.startswith('NN'):
            color = "white"
        elif tag.startswith('VB'):
            color = "green"
        elif tag.startswith('JJ'):
            color = "orange"
        elif tag.startswith('RB'):
            color = "purple"
        elif tag.startswith('DT') or tag.startswith('IN'):
            color = "red"
        formatted_text.append(f"<b><span style='color:{color}'>{word} ({tag})</span></b>")
    return ' '.join(formatted_text)

# Formatar e exibir a sada
from IPython.core.display import HTML
formatted_text = format_pos_tags(pos_tags)
HTML(formatted_text)
```

**Objetivo:** Compreender e aplicar o processo de stemming.

**Tarefa 3.1:** Utilizar o algoritmo de RSLP para *stemming* em um conjunto de palavras fornecidas.

**Definio: Stemming**

O stemming  uma tcnica fundamental no Processamento de Linguagem Natural que tem como objetivo reduzir palavras flexionadas ou derivadas ao seu radical ou raiz comum. Essa tcnica  frequentemente utilizada para normalizar texto, reduzir a dimensionalidade do vocabulrio e melhorar a eficincia de algoritmos de anlise de texto.

**Objetivos do Stemming:**


*   *Reduo do Vocabulrio:* Diminuir o nmero de formas de palavras diferentes tratadas pelos modelos de Processamento de Linguagem Natural .

*  *Normalizao de Texto:* Unificar diferentes formas de uma palavra para melhorar a consistncia do texto.

*   *Melhoria de Desempenho:* Reduzir a carga computacional ao diminuir a diversidade lexical sem perder muito do significado semntico.


**Algoritmos de Stemming:**


1.   *Algoritmo de Porter:* Um dos algoritmos mais populares, desenvolvido por Martin Porter em 1980. Opera atravs de uma srie de regras de reescrita aplicadas sequencialmente. Exemplo: "correndo"  "corr", "facilmente"  "facil".

2.   *Algoritmo de Snowball (Porter2):* Uma verso aprimorada do algoritmo de Porter. Oferece maior preciso e manipulao de excees. Exemplo: "correu"  "corr", "facilmente"  "facil".

**Implementao em Python:**
A biblioteca NLTK (Natural Language Toolkit) fornece implementaes dos algoritmos de stemming, facilitando sua aplicao prtica. No entanto, para a lngua portuguesa, podemos utilizar a biblioteca RSLPStemmer do NLTK, que  mais adequada para este idioma:

1.   execute o cdigo e analise a sada; e
2.   mude as palavras das linhas 30 e 50 do cdigo Python e re-execute.

```python
import subprocess
import sys

# Funo para instalar pacotes
def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# Verificar se o NLTK est instalado
try:
    import nltk
except ImportError:
    print("NLTK no est instalado. Instalando...")
    install("nltk")
    import nltk

# Configurar o diretrio de dados do NLTK no Colab
nltk.data.path.append('/root/nltk_data')    # add: linha adicionada

# Verificar se o stemmer RSLP est instalado
try:
    nltk.data.find('stemmers/rslp')
except LookupError:
    print("Stemmer RSLP no est instalado. Instalando...")
    nltk.download('rslp')

# Agora, o cdigo para stemmizar as palavras em portugus
from nltk.stem import RSLPStemmer

# Instanciando o RSLP Stemmer
rslp_stemmer = RSLPStemmer()

# Lista de palavras em portugus para stemmizar
words_pt = ['correndo', 'correu', 'correm', 'facilmente', 'justamente']

# Aplicando o RSLP Stemmer
stems_rslp = [rslp_stemmer.stem(word) for word in words_pt]
print("Stemming em portugus (RSLP):", stems_rslp)

# Verificar se o SnowballStemmer est instalado
try:
    nltk.data.find('stemmers/snowball')
except LookupError:
    print("Stemmer Snowball no est instalado. Instalando...")
    nltk.download('snowball_data')

# Agora, o cdigo para stemmizar as palavras em ingls usando Snowball (Porter2)
from nltk.stem import SnowballStemmer

# Instanciando o Snowball Stemmer para ingls
snowball_stemmer = SnowballStemmer("english")

# Lista de palavras em ingls para stemmizar
words_en = ['running', 'ran', 'runs', 'easily', 'fairly']

# Aplicando o Snowball Stemmer
stems_snowball = [snowball_stemmer.stem(word) for word in words_en]
print("Stemming em ingls (Snowball):", stems_snowball)
```

**NOTA:** **Limitaes do Stemming:**

1.  *Reduo Excessiva:* Pode resultar em radicais que no so palavras vlidas, o que pode afetar a interpretao semntica.
2.   *Sensibilidade ao Idioma:* Os algoritmos geralmente so projetados para um idioma especfico e podem no ser eficazes em outros.

# **Atividade 4: Comparao entre Stemming e Lematizao**

**Objetivo:** Comparar os processos de stemming e lematizao.

**Relembrando as tcnias:**

**de Lematizao:** A lematizao  uma tcnica de processamento de linguagem natural que transforma palavras flexionadas na sua forma base ou "lema", considerando o contexto e as regras lingusticas. Por exemplo, "correndo", "correu" e "correm" so reduzidos ao lema "correr".

**de Stemming:** O stemming  uma tcnica que reduz palavras flexionadas ou derivadas ao seu radical ou raiz comum, sem considerar o contexto lingustico. Por exemplo, "correndo", "correu" e "correm" so reduzidos ao radical "corr".

```python
import subprocess
import sys

# Funo para instalar um pacote se ele no estiver instalado
def install_package(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# Verificar e instalar o NLTK se necessrio (Stemming em Portugus)
try:
    import nltk
except ImportError:
    install_package('nltk')
    import nltk

try:
    import spacy
except ImportError:
    install_package('spacy')
    import spacy

import nltk
from nltk.stem import WordNetLemmatizer, RSLPStemmer

# Baixar recursos necessrios para lematizao e stemming
nltk.download('rslp')

# Verificar e instalar o modelo de portugus do spaCy
try:
    nlp = spacy.load("pt_core_news_sm")
except Exception:
    subprocess.check_call([sys.executable, "-m", "spacy", "download", "pt_core_news_sm"])
    nlp = spacy.load("pt_core_news_sm")

# Inicializar o lematizador e o stemmer
lemmatizer = WordNetLemmatizer()
stemmer = RSLPStemmer()

# Lista de palavras para lematizao e stemming
palavras = ['correndo', 'correu', 'correm', 'facilmente', 'justamente']

# Lematizao das palavras utilizando o spaCy
def lemmatize_word(word):
    doc = nlp(word)
    # Retorna o lema do primeiro token
    return doc[0].lemma_

# Lematizao das palavras
# Lematizao  o processo de reduzir uma palavra  sua forma base ou raiz
palavras_lematizadas = [lemmatize_word(palavra) for palavra in palavras]

# Stemming das palavras
# Stemming  o processo de reduzir uma palavra ao seu radical
palavras_stemmed = [stemmer.stem(palavra) for palavra in palavras]

# Comparao dos resultados de lematizao e stemming
# Cria uma lista de tuplas contendo a palavra original, a palavra lematizada e a palavra aps stemming
comparacao = list(zip(palavras, palavras_lematizadas, palavras_stemmed))

# Define o tamanho mximo para cada coluna
col_width = 20
# Imprime o cabealho da tabela com espaamento fixo
print(f"{'Palavra':<{col_width}}{'Palavra Lematizada':<{col_width}}{'Palavra_stemmed':<{col_width}}")
print("-" * (col_width * 3))  # Linha separadora

# Itera sobre cada tupla e imprime seus valores formatados com espaamento fixo
for tupla in comparacao:
    print(f"{tupla[0]:<{col_width}}{tupla[1]:<{col_width}}{tupla[2]:<{col_width}}")
```

**NOTA: Comparao *Stemming* vs. Lematizao:**

1.   *Stemming:* Reduz palavras ao seu radical de forma mecnica, sem levar em considerao o contexto lingustico.
2.  *Lematizao:* Considera o contexto e transforma as palavras na sua forma base ou lema.

Cdigo de Implementao
Aqui est o cdigo que implementa a classificao das sentenas:

**Classificao de Sentenas em Portugus usando spaCy**

Neste exemplo, apresentamos como utilizar a biblioteca spaCy para classificar sentenas em portugus como "positivo", "negativo" ou "neutro" com base na presena de palavras-chave. Vamos explorar o cdigo passo a passo para entender seu funcionamento.

**Funo de Classificao de Sentenas**

A seguir, definimos uma funo chamada classify_sentence que classifica uma sentena como "positivo", "negativo" ou "neutro" com base nas palavras-chave encontradas na frase:

**Tarefa 5.1:** Dado o cdigo do caso de uso de classificao de texto em questo:
1.   execute o cdigo e analise a sada; e
2.   mude as frases das linhas 19 at 22 do cdigo Python e re-execute.


```python
import spacy

# Carregar o modelo em portugus do spaCy
nlp = spacy.load('pt_core_news_sm')

# Funo para classificar sentenas como "positivo" ou "negativo"
def classify_sentence(sentence):
    doc = nlp(sentence)
    # Verificar a presena de palavras-chave
    if any(token.lemma_ in ['timo', 'excelente'] for token in doc):
        return "positivo"
    elif any(token.lemma_ in ['ruim', 'pssimo', 'terrvel'] for token in doc):
        return "negativo"
    else:
        return "neutro"

# Exemplos de sentenas para classificao
sentences = [
    "Este restaurante  excelente, adorei a comida!",
    "O atendimento nesta loja  pssimo, no recomendo.",
    "O filme foi bom, mas o final deixou a desejar.",
    "A viagem foi tima, tudo correu conforme o planejado."
]

# Classificar e imprimir os resultados
for sentence in sentences:
    classification = classify_sentence(sentence)
    print(f"Sentena: '{sentence}' - Classificao: {classification}")
```

**Atividade 6: Algoritmo de Porter para stemming**

**Objetivo:** Explorar algoritmos e tcnicas comuns na anlise morfolgica.

**Como Funciona o Algoritmo de Porter?** O algoritmo de Porter aplica uma srie de regras para transformar palavras em suas razes. Aqui esto os principais passos:

1.   Remoo de Sufixos Plurais e de Gerndio: O algoritmo comea removendo sufixos comuns que indicam pluralidade ou gerndio. Por exemplo, "correndo" pode ser transformado em "corr".
2.   Tratamento de Sufixos Derivacionais: Em seguida, remove sufixos que so usados para formar novas palavras a partir de outras, como "mente" em "facilmente". Isso transformaria "facilmente" em "facil".
3.   Simplificao de Palavras: Aplica regras para simplificar palavras, removendo sufixos adicionais e padronizando as formas.
4.   Remoo de Sufixos Finais: Por fim, remove quaisquer sufixos restantes que possam ser desnecessrios para a raiz da palavra.

```python
def porter_stemmer(word):
    """
    Implementao simplificada do algoritmo de Porter para stemming.
    """
    suffixes = [
        ('ational', 'ate'), ('tional', 'tion'), ('enci', 'ence'), ('anci', 'ance'),
        ('izer', 'ize'), ('bli', 'ble'), ('alli', 'al'), ('entli', 'ent'),
        ('eli', 'e'), ('ousli', 'ous'), ('ization', 'ize'), ('ation', 'ate'),
        ('ator', 'ate'), ('alism', 'al'), ('iveness', 'ive'), ('fulness', 'ful'),
        ('ousness', 'ous'), ('aliti', 'al'), ('iviti', 'ive'), ('biliti', 'ble'),
        ('logi', 'log')
    ]

    def replace_suffix(word, suffixes):
        for suffix, replacement in suffixes:
            if word.endswith(suffix):
                return word[:-len(suffix)] + replacement
        return word

    def step_1a(word):
        if word.endswith('sses'):
            return word[:-2]
        elif word.endswith('ies'):
            return word[:-2]
        elif word.endswith('ss'):
            return word
        elif word.endswith('s'):
            return word[:-1]
        return word

    def step_1b(word):
        if word.endswith('eed'):
            return word[:-1] if len(word) > 3 else word
        elif word.endswith('ed') and has_vowel(word[:-2]):
            return replace_suffix(word[:-2], suffixes)
        elif word.endswith('ing') and has_vowel(word[:-3]):
            return replace_suffix(word[:-3], suffixes)
        return word

    def has_vowel(word):
        for char in word:
            if char in 'aeiou':
                return True
        return False

    def step_2(word):
        return replace_suffix(word, suffixes)

    word = step_1a(word)
    word = step_1b(word)
    word = step_2(word)
    return word

# Exemplo de uso
words = ["running", "talked", "processes", "organizational"]
stemmed_word = [porter_stemmer(word) for word in words]
print(f"Palavra original: {words}")
print(f"Palavra reduzida (stemming): {stemmed_word}")
```

# **Atividade 7: Problema de Indexao de pequenos documentos**




##**Indexao usando a abordagem TF-IDF**

No campo do Processamento de Linguagem Natural e da Recuperao de Informao, o problema de indexao de documentos  crucial para organizar e recuperar informaes de grandes volumes de texto de forma eficiente. Um mtodo amplamente utilizado para esse fim  o TF-IDF, que significa "Term Frequency-Inverse Document Frequency". Este mtodo ajuda a medir a importncia de uma palavra em um documento dentro de uma coleo ou corpus de documentos.

## Cenrio

Imagine que voc tem uma grande coleo de documentos, como artigos cientficos, posts de blogs, ou pginas da web. Quando um usurio faz uma consulta de busca, o sistema precisa rapidamente identificar quais documentos so mais relevantes para essa consulta. O problema de indexao de documentos , portanto, criar uma estrutura que permita essa recuperao eficiente e precisa de informaes.


## O Que  TF-IDF?

TF-IDF  uma mtrica que ajuda a avaliar a importncia de uma palavra em um documento, levando em conta tanto a frequncia da palavra no documento (TF) quanto a frequncia dessa palavra em todo o corpus de documentos (IDF). Vamos entender cada componente:


### TF (Term Frequency)

Frequncia do termo no documento. Calcula-se quantas vezes uma palavra aparece em um documento em relao ao nmero total de palavras nesse documento. A frmula :

$$
\text{TF}(t, d) = \frac{f_{t,d}}{N_d}
$$

- $(f_{t,d})$   a contagem de ocorrncias do termo \( t \) no documento \( d \)
- $(N_d)$   o nmero total de termos no documento \( d \)


### IDF (Inverse Document Frequency)

Frequncia inversa do documento. Avalia a importncia do termo no corpus inteiro, penalizando termos comuns em muitos documentos. A frmula :

$$
\text{IDF}(t, D) = \log \left( \frac{N}{|d \in D: t \in d|} \right)
$$

- $( N )$  o nmero total de documentos no corpus
- $( |d \in D: t \in d| )\$  o nmero de documentos onde o termo \( t \) aparece


### TF-IDF

Produto do TF e IDF. Combina ambas as mtricas para calcular a importncia do termo no documento, ajustada pela frequncia no corpus. A frmula :

$$
\text{TF-IDF}(t, d, D) = \text{TF}(t, d) \times \text{IDF}(t, D)
$$

## Como o TF-IDF Resolve o Problema de Indexao

- **Relevncia de Termos**: TF-IDF ajuda a identificar palavras importantes para um documento especfico, reduzindo a influncia de termos comuns e irrelevantes.
- **Ranking de Documentos**: Quando uma busca de documento  feita, o sistema pode calcular o TF-IDF de cada termo da consulta em cada documento e somar esses valores para determinar quais documentos so mais relevantes.
- **Eficincia na Busca**: Utilizando estruturas como matrizes esparsas, os sistemas podem rapidamente calcular e comparar valores de TF-IDF, permitindo uma recuperao eficiente de documentos relevantes.

## Exemplo Prtico

Considere o seguinte corpus com cinco documentos:

- Documento 1: "O gato est correndo no jardim."
- Documento 2: "Os gatos gostam de correr e brincar."

- Documento 5: "O gado preto da Renata dorme na varanda."

Vamos calcular o TF-IDF para a palavra "gosta".

### TF (Term Frequency)

- Documento 1: "gosta" no aparece (TF = 0)
- Documento 2: "gosta" aparece uma vez em 6 palavras (TF = 1/6  0.167)
- Documento 3: "gosta" aparece uma vez em 6 palavras (TF = 1/6  0.167)
- Documento 4: "gosta" no aparece (TF = 0)
- Documento 5: "gosta" no aparece (TF = 0)

### IDF (Inverse Document Frequency)

- "gosta" aparece em 2 dos 5 documentos.
- IDF = $$ \log \left( \frac{5}{2} \right) \approx 0.398 $$

### TF-IDF

- Documento 1: 0 * 0.398 = 0
- Documento 2: 0.167 * 0.398  0.066
- Documento 3: 0.167 * 0.398  0.066
- Documento 4: 0 * 0.398 = 0
- Documento 5: 0 * 0.398 = 0

### Clculo da Similaridade usando o Coeficiente do Cosseno

A similaridade entre dois documentos pode ser calculada usando o coeficiente do cosseno, que mede o ngulo entre dois vetores. A frmula para calcular a similaridade do cosseno entre dois vetores \($\mathbf{A} $\) e \($\mathbf{B}$\) :

$$
\text{similaridade}_{\cos}(\mathbf{A}, \mathbf{B}) = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}
$$

onde:

- \( $\mathbf{A} \cdot \mathbf{B} $\)  o produto escalar dos vetores \( $\mathbf{A}$ \) e \( $\mathbf{B}$ \)
- \( $\|\mathbf{A}\|$ \)  a norma (magnitude) do vetor \( $\mathbf{A} $\)
- \( $\|\mathbf{B}\|$ \)  a norma (magnitude) do vetor \( $\mathbf{B} $\)



### Exemplo de Clculo da Similaridade do Cosseno

Vamos calcular a similaridade do cosseno entre os Documentos 2 e 3 usando seus vetores TF-IDF simplificados. Suponha que temos os seguintes vetores TF-IDF para simplificao:


-\($\mathbf{A} = [0.066, 0, 0, 0, 0]$\) (Documento 2)
-\($\mathbf{B} = [0.066, 0, 0, 0, 0]$\) (Documento 3)


A similaridade do cosseno  calculada como:

$
\text{similaridade}_{\cos}(\mathbf{A}, \mathbf{B}) = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|} = \frac{0.066 \times 0.066}{\sqrt{0.066^2} \times \sqrt{0.066^2}} = \frac{0.004356}{0.066 \times 0.066} = 1$

Neste caso, a similaridade do cosseno entre os Documentos 2 e 3  1, indicando que eles so idnticos em termos de contedo de TF-IDF.

### Resultados

A importncia da palavra "gosta"  maior nos Documentos 2 e 3, ambos com um valor TF-IDF de aproximadamente 0.066. Os Documentos 1, 4 e 5 tm um valor TF-IDF de 0 para "gosta", indicando que a palavra no aparece nesses documentos.

```python
# Etapa 1: Verificao e Instalao das Bibliotecas Necessrias
import subprocess
import sys

def install_and_import(package):
    try:
        __import__(package)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# Instalar as bibliotecas necessrias
install_and_import('spacy')
install_and_import('nltk')
install_and_import('scikit-learn')
install_and_import('pandas')

# Baixar o modelo de portugus do spaCy, caso no esteja instalado
subprocess.run([sys.executable, "-m", "spacy", "download", "pt_core_news_sm"])

# Etapa 2: Importar as Bibliotecas e Inicializar Recursos
import spacy
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Baixar recursos necessrios do nltk para stemming e lematizao
nltk.download('rslp')  # Baixar o stemmer RSLP para portugus
nltk.download('wordnet')  # Baixar o WordNet para lematizao
nltk.download('omw-1.4')  # Baixar o Open Multilingual Wordnet
nltk.download('punkt')  # Baixar o tokenizer 'punkt' para portugus

from nltk.stem import RSLPStemmer
from nltk.stem import WordNetLemmatizer

# Carregar o modelo de portugus do spaCy
nlp = spacy.load("pt_core_news_sm")

# Inicializar o lematizador e o stemmer
lemmatizer = WordNetLemmatizer()
stemmer = RSLPStemmer()

# Etapa 3: Definir Funes de Pr-Processamento

# Funo de lematizao usando spaCy
# Esta funo utiliza o spaCy para lematizar o texto, retornando os lemas dos tokens.
def lematizar(texto):
    doc = nlp(texto)
    return ' '.join([token.lemma_ for token in doc])

# Funo de stemming usando NLTK
# Esta funo utiliza o nltk para aplicar stemming aos tokens do texto.
def aplicar_stemming(texto):
    tokens = nltk.word_tokenize(texto, language='portuguese')
    return ' '.join([stemmer.stem(token) for token in tokens])

# Funo de pr-processamento completo
# Esta funo aplica lematizao seguida de stemming ao texto.
def preprocessamento(texto):
    texto_lematizado = lematizar(texto)
    texto_stemmed = aplicar_stemming(texto_lematizado)
    return texto_stemmed

# Etapa 4: Pr-Processar Documentos e Calcular TF-IDF

# Lista de documentos de exemplo para indexao
documentos = [
    "O gato est correndo no jardim.",
    "Os gatos gostam de correr e brincar.",
    "Ela gosta de estudar lingustica computacional.",
    "Lingustica computacional  uma rea fascinante.",
    "O gato preto da Renata dorme na varanda."
]

# Aplicar pr-processamento em todos os documentos
# Esta linha aplica a funo de pr-processamento em cada documento.
documentos_preprocessados = [preprocessamento(doc) for doc in documentos]

# Exibir os documentos preprocessados
print("Documentos Preprocessados:")
for doc in documentos_preprocessados:
    print(doc)

# Inicializar o vetor TF-IDF
vectorizer = TfidfVectorizer()

# Ajustar e transformar os documentos preprocessados
# Esta linha ajusta e transforma os documentos preprocessados em uma matriz TF-IDF.
tfidf_matrix = vectorizer.fit_transform(documentos_preprocessados)

# Obter os termos (features) do TF-IDF
termos = vectorizer.get_feature_names_out()

# Exibir a matriz TF-IDF usando um DataFrame para melhor visualizao
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=termos)

# Exibir a matriz TF-IDF
print("Matriz TF-IDF:")
print(df_tfidf)

# Etapa 5: Consulta de Documentos e Calcular Similaridade

# Funo para processar consulta e calcular similaridade
# Esta funo pr-processa a consulta, transforma em vetor TF-IDF e
# calcula a similaridade coseno.
def consultar_documentos(consulta):
    consulta_preprocessada = preprocessamento(consulta)
    consulta_tfidf = vectorizer.transform([consulta_preprocessada])
    similaridades = cosine_similarity(consulta_tfidf, tfidf_matrix)
    documento_mais_similar = similaridades.argmax()
    return documento_mais_similar, similaridades

# Exemplo de consultas
consultas = [
    "gato no jardim",
    "estudar lingustica",
    "gato da Renata"
]

# Processar cada consulta e exibir resultados
for consulta in consultas:
    indice, similaridades = consultar_documentos(consulta)
    print(f"Consulta: '{consulta}'")
    print(f"Documento mais similar: {documentos[indice]}")
    print(f"Similaridades: {similaridades}\n")
```
